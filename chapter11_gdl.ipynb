```python
import numpy as np
from collections import defaultdict
import random

class Word2Vec:
    def __init__(self, corpus, embedding_dim=100, window_size=2, negative_samples=5, learning_rate=0.025):
        self.corpus = corpus
        self.embedding_dim = embedding_dim
        self.window_size = window_size
        self.negative_samples = negative_samples
        self.learning_rate = learning_rate
        self.word2id = defaultdict(lambda: len(self.word2id))
        self.id2word = {}
        self.vocab_size = 0
        self.word_counts = {}
        self.total_word_count = 0
        self.embeddings = None
        self.init_vocab()
        self.init_embeddings()

    def init_vocab(self):
        for sentence in self.corpus:
            for word in sentence.split():
                self.word_counts[word] = self.word_counts.get(word, 0) + 1
                self.total_word_count += 1
                self.word2id[word]

        self.vocab_size = len(self.word2id)
        self.id2word = {v: k for k, v in self.word2id.items()}

    def init_embeddings(self):
        self.embeddings = (np.random.rand(self.vocab_size, self.embedding_dim) - 0.5) / self.embedding_dim
        print(self.embeddings)

    def train(self, epochs=5):
        for epoch in range(epochs):
            loss = 0.0
            for sentence in self.corpus:
                tokens = sentence.split()
                for i, target_word in enumerate(tokens):
                    for j in range(max(0, i - self.window_size), i + self.window_size + 1):
                        if j != i and j >= 0 and j < len(tokens):
                            context_word = tokens[j]
                            print('context_word:',context_word)
                            print('')
                            loss += self.train_pair(target_word, context_word)

            print('Epoch:', epoch + 1, 'Loss:', loss)

    def train_pair(self, target_word, context_word):
        target_id = self.word2id[target_word]
        context_id = self.word2id[context_word]
        negative_ids = [self.get_negative_sample(target_id) for i in range(self.negative_samples)]
        embeddings = np.vstack((self.embeddings[context_id], self.embeddings[negative_ids]))
        output = self.softmax(np.dot(embeddings, self.embeddings[target_id]))
        error = output[0] - 1
        gradient = error * self.learning_rate
        self.embeddings[target_id] -= gradient * np.sum(embeddings, axis=0)
        for id in [context_id] + negative_ids:
            self.embeddings[id] -= gradient * self.embeddings[target_id]

        return -np.log(output[0])

    def get_negative_sample(self,target_id):
        while True:
            sample_id = random.randint(0, self.vocab_size - 1)
            if sample_id != target_id:
                return sample_id

    def softmax(self, x):
        e_x = np.exp(x - np.max(x))
        return e_x / e_x.sum()

    def get_word_vector(self, word):
        return self.embeddings[self.word2id[word]]
# Define the corpus
# corpus = ["The quick brown fox jumps over the lazy dog.",
#           "She sells seashells by the seashore.",
#           "Peter Piper picked a peck of pickled peppers."]
corpus = [
    
"2022-03-27 12:00:01 192.168.1.101 203.0.113.5 TCP Allowed",
"2022-03-27 12:00:03 10.1.2.3 172.16.8.10 UDP Blocked",
"2022-03-27 12:00:05 192.168.2.75 8.8.8.8 TCP Allowed",
"2022-03-27 12:00:07 10.0.0.123 192.168.15.30 ICMP Allowed",
"2022-03-27 12:00:09 172.30.5.12 192.168.50.100 UDP Allowed",
"2022-03-27 12:00:11 192.168.80.12 172.16.2.35 TCP Blocked",
"2022-03-27 12:00:13 10.0.20.5 192.168.4.20 UDP Allowed",
"2022-03-27 12:00:15 192.168.100.2 172.30.10.5 ICMP Blocked",
"2022-03-27 12:00:17 172.16.5.8 192.168.10.15 TCP Allowed",
"2022-03-27 12:00:19 192.168.200.25 172.30.15.80 UDP Allowed",
"2022-03-27 12:00:21 10.0.30.50 172.16.20.30 TCP Blocked",
"2022-03-27 12:00:23 192.168.150.10 8.8.4.4 ICMP Allowed",
"2022-03-27 12:00:25 172.16.25.40 192.168.5.5 UDP Blocked",
"2022-03-27 12:00:27 192.168.60.5 10.1.1.1 TCP Allowed",
"2022-03-27 12:00:29 172.16.30.40 192.168.3.20 UDP Allowed",
"2022-03-27 12:00:31 192.168.250.50 172.30.8.10 TCP Blocked",
"2022-03-27 12:00:33 172.16.1.10 10.0.40.5 UDP Allowed",
"2022-03-27 12:00:35 192.168.175.20 172.16.50.15 ICMP Allowed",
"2022-03-27 12:00:37 10.1.20.40 192.168.70.5 TCP Blocked"
]
w2v = Word2Vec(corpus=corpus)
w2v.train(epochs=10)
print(w2v.get_word_vector('10.1.20.40'))
```
